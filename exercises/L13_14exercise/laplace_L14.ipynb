# Part 1

set parameters

```python
NX    = 10
NY    = 10
XMIN  = 0
XMAX  = 1
YMIN  = 0
YMAX  = 1
NITER = 50  # max number of iterations
```

Initialize the grids

```python
u = np.zeros(shape=(NX,NY))
```

set initial conditions (assume 0 everywhere)

```python
def initial(u,nx=NX,ny=NY):
    u[:,:] = 0
    return u
```

Boundary conditions

```python
def set_boundary_conditions(u,nx=NX,ny=NY):
    """
    Input: u[i][j]
    
    Output: u[i][j]
    
    B.C.: top    (y=1): u = 1
          button (y=0): u = 0
          left   (x=0): u = 0
          right  (x=1): u = 0
    """

    # TODO: set the boundary conditions here
    u[:,ny-1] = 1
    u[:,0] = 0    
    u[0,:] = 0
    u[nx-1,:] = 0
    
    return u
```

## Jacobi method
Do one iteration

```python
def evolve_jacobi(u,nx=NX,ny=NY):
    """
    do one Jacobi iteration
    
    Inputs: u[size of NX][size of NY]
    NX: number of points in x-axis
    NY: number of points in y-axis
    
    Outputs: u
    
    """
    uold = u.copy()
    set_boundary_conditions(u, nx=NX, ny=NY)
    for i in range(1,nx-1):
        for j in range(1,ny-1):
            u[i,j] = 0.25*(uold[i+1,j]+uold[i-1,j]+uold[i,j+1]+uold[i,j-1])

    return u
```

Check convergence by comparing the differecne between u1 and u2

```python
def diff(u1,u2,nx=NX,ny=NY):
    error = 0.0
    for i in range(nx):
        for j in range(ny):
            error += abs(u1[i,j] - u2[i,j])
    return error
```

Start iterations

```python
u = initial(u)
errors = np.zeros(NITER)
for n in range(NITER):
    
    plt.figure(1,figsize=(12,5))
    plt.subplot(121)
    plt.title("N= {}".format(n))
    clear_output(wait=True)
    
    # make a copy of the u in the previous step
    uold = u.copy()
    
    # do one iteration
    u = evolve_jacobi(u)
    
    # check the difference
    err = diff(uold,u)
    errors[n] = err
    
    # plot the results
    plt.imshow(u.T,origin='lower',extent=[XMIN,XMAX,YMIN,YMAX],interpolation='none')
    plt.colorbar()
    plt.subplot(122)
    plt.plot(errors,'k-')
    plt.xlim([0,NITER+1])
    plt.xlabel("N")
    plt.yscale('log')
    #time.sleep(0.0001)
    plt.show()
```

## Gauss-Seidel method

```python
def evolve_gauss_seidel(u,nx=NX,ny=NY):
    """
    do one Gauss-Seidel iteration
    """
    set_boundary_conditions(u, nx=NX, ny=NY)
    for i in range(1,nx-1):
        for j in range(1,ny-1):
            u[i,j] = 0.25*(u[i+1,j]+u[i-1,j]+u[i,j+1]+u[i,j-1])
            
    return u
```

## Successive over-relax method:

w < 1 : under-relaxation
w = 1 : Gauss-Seidel method
w > 1 : over-relaxation

```python
def evolve_successive_over_relax(u,nx=NX,ny=NY):
    
    """
    do Successive over-relax iteration
    """
    
    w = 1.2
    uold = u.copy()
    uGS = evolve_gauss_seidel(u)
    set_boundary_conditions(u, nx=NX, ny=NY)
    u = (1-w)*uold+w*uGS
    
    return u
```

```python
#TODO: Visualize the results here
u = initial(u)
errors = np.zeros(NITER)
for n in range(NITER):
    
    plt.figure(1,figsize=(12,5))
    plt.subplot(121)
    plt.title("N= {}".format(n))
    clear_output(wait=True)
    
    # make a copy of the u in the previous step
    uold = u.copy()
    
    # do one iteration
    u = evolve_successive_over_relax(u)
    
    # check the difference
    err = diff(uold,u)
    errors[n] = err
    
    # plot the results
    plt.imshow(u.T,origin='lower',extent=[XMIN,XMAX,YMIN,YMAX],interpolation='none')
    plt.colorbar()
    plt.subplot(122)
    plt.plot(errors,'k-')
    plt.xlim([0,NITER+1])
    plt.xlabel("N")
    plt.yscale('log')
    #time.sleep(0.0001)
    plt.show()
```

## Multi-grid method

```python
NX1   = 25
NY1   = 25
NX2   = 50
NY2   = 50
NX3   = 100
NY3   = 100
NITER_MG = 100
u1 = np.zeros(shape=(NX1,NY1))
u2 = np.zeros(shape=(NX2,NY2))
u3 = np.zeros(shape=(NX3,NY3))
```

```python
def set_boundary_conditions_MG(u,nx,ny):
    """
    Input: u[i][j]
    
    Output: u[i][j]
    
    B.C.: top    (y=1): u = 1
          button (y=0): u = 0
          left   (x=0): u = 0
          right  (x=1): u = 0
    """

    # TODO: set the boundary conditions here
    u[:,ny-1] = 1
    u[:,0] = 0    
    u[0,:] = 0
    u[nx-1,:] = 0
    
    return u
```

```python
def evolve_gauss_seidel_MG(u,nx,ny):
    """
    do one Gauss-Seidel iteration
    """
    set_boundary_conditions_MG(u, nx, ny)
    for i in range(1,nx-1):
        for j in range(1,ny-1):
            u[i,j] = 0.25*(u[i+1,j]+u[i-1,j]+u[i,j+1]+u[i,j-1])
            
    return u
```

```python
def restriction(u_fine,u_coarser,NXC,NYC):
    """
    from fine grid to coarser grid.
    Assume the gird size is different by a factor of 2
    
    update the coarser grid based on information of the fine grid
    
    Inputs:
    
    u_fine: size of 2 NXC and 2 NYC
    u_coarser: size of NXC and NYC
    
    
    Outputs:
    
    u: size of NXC and NYC
    
    
    """
    for i in range(NXC):
        for j in range(NYC):
            ii = 2*i
            jj = 2*j
            u_coarser[i,j] = 0.25*(u_fine[ii,jj]+u_fine[ii+1,jj]+u_fine[ii,jj+1]+u_fine[ii+1,jj+1])
    
    uout = u_coarser
    
    return uout  
```

```python
def prolongation(u_fine, u_coarser,NXC, NYC):
    """
    from coarser grid to fine grid
    
    update the fine grid based on information of the coarser grid
    
    Inputs:
    
    u_fine: size of 2 NXC and 2 NYC
    u_coarser: size of NXC and NYC
    
    
    Outputs:
    
    u: size of 2 NXC and 2 NYC
    
    """

    # To simplify the problem, we could do direct mapping without interpolation
    
    for i in range(NXC):
        for j in range(NYC):
            ii = 2*i
            jj = 2*j
            
            u_fine[ii, jj] = u_coarser[i,j]
            u_fine[jj+1, jj] = u_coarser[i,j]
            u_fine[ii, jj+1] = u_coarser[i,j]
            u_fine[ii+1, jj+1] = u_coarser[i,j]
    
    uout = u_fine
    return uout
```

```python
#TODO: To the full multigird cycle here and Visualize the results
def smoother(u,nx,ny):
    for n in range(NITER_MG):
        u = evolve_gauss_seidel_MG(u,nx,ny)
    return u
```

```python
def smoother_show(u,nx,ny):
    errors = np.zeros(NITER_MG)

    for n in range(NITER_MG):
    
        plt.figure(1,figsize=(12,5))
        plt.subplot(121)
        plt.title("N= {}".format(n))
        clear_output(wait=True)
    
        # make a copy of the u in the previous step
        uold = u.copy()
    
        # do one iteration
        u = evolve_gauss_seidel_MG(u,nx,ny)
    
        # check the difference
        err = diff(uold,u)
        errors[n] = err
    
        # plot the results
        plt.imshow(u.T,origin='lower',extent=[XMIN,XMAX,YMIN,YMAX],interpolation='none')
        plt.colorbar()
        plt.subplot(122)
        plt.plot(errors,'k-')
        plt.xlim([0,NITER_MG+1])
        plt.xlabel("N")
        plt.yscale('log')
        #time.sleep(0.0001)
        plt.show()
        
    return u
```

```python
u3 = initial(u3)
u2 = initial(u2)
u1 = initial(u1) 
    
u3 = smoother(u3,NX3,NY3)
u2 = restriction(u3,u2,NX2,NY2)

u2 = smoother(u2,NX2,NY2)
u1 = restriction(u2,u1,NX1,NY1)

u1 = smoother(u1,NX1,NY1)
u2 = prolongation(u2,u1,NX1,NY1)

u2 = smoother(u2,NX2,NY2)
u3 = prolongation(u3,u2,NX2,NY2)

u3 = smoother_show(u3,NX3,NY3)
```
